```python
usuarios = {
    "admin": {"password": "admin123", "empresa": "todas"},
    "juan": {"password": "1234", "empresa": "Colmena"},
    "maria": {"password": "1234", "empresa": "Almasa"},
    "carlos": {"password": "1234", "empresa": "GyJ Ferreterias"}
}

login_html = """
<!DOCTYPE html>
<html>
<head>
  <title>Login</title>
</head>
<body style="text-align:center; font-family:Arial;">

<h2>Iniciar Sesion</h2>

<form method="POST">
  <input type="text" name="usuario" placeholder="Usuario" required><br><br>
  <input type="password" name="password" placeholder="Contraseña" required><br><br>

  <select name="empresa">
    {% for emp in empresas %}
      <option value="{{emp}}">{{emp}}</option>
    {% endfor %}
  </select><br><br>

  <button type="submit">Ingresar</button>
</form>

<p style="color:red;">{{error}}</p>

</body>
</html>
"""

dashboard_html = """
<!DOCTYPE html>
<html>
<head>
  <title>Panel</title>

  <style>
    body {
        margin: 0;
        font-family: Arial;
    }

    .navbar {
        background-color: #2c3e50;
        overflow: hidden;
        display: flex;
        justify-content: space-between;
        padding: 10px;
        color: white;
    }

    .navbar h2 {
        margin: 0;
    }

    .menu {
        display: flex;
        gap: 20px;
    }

    .menu a {
        color: white;
        text-decoration: none;
        cursor: pointer;
    }

    .menu a:hover {
        text-decoration: underline;
    }

    .contenido {
        padding: 40px;
        text-align: center;
    }
  </style>

</head>

<body>

<div class="navbar">
    <h2>{{empresa}}</h2>

<div class="menu">
    <a href="/mercadeo?usuario={{usuario}}&empresa={{empresa}}">Mercadeo</a>
    <a href="/talento?usuario={{usuario}}&empresa={{empresa}}">Talento</a>
    <a href="/gerencia?usuario={{usuario}}&empresa={{empresa}}">Gerencia</a>
    <a href="/">Cerrar sesión</a>
</div>
</div>

<div class="contenido" id="contenido">
    <h1>Bienvenido {{usuario}}</h1>
</div>

<script>
function cargarModulo(modulo){
    let contenido = document.getElementById("contenido");

    if(modulo === "mercadeo"){
        contenido.innerHTML = "<h2>Módulo de Mercadeo</h2>";
    }

    if(modulo === "talento"){
        contenido.innerHTML = "<h2>Talento Humano</h2>";
    }

    if(modulo === "gerencia"){
        contenido.innerHTML = "<h2>Gerencia</h2>";
    }
}

function cerrarSesion(){
    window.location.href = "/";
}
</script>

</body>
</html>
"""

mercadeo_html = """
<!DOCTYPE html>
<html>
<head>
<title>Mercadeo</title>

<style>
body {
    margin: 0;
    font-family: Arial;
}

.navbar {
    background-color: #34495e;
    padding: 10px;
    display: flex;
    gap: 20px;
}

.navbar a {
    color: white;
    text-decoration: none;
    cursor: pointer;
}

.navbar a:hover {
    text-decoration: underline;
}

.contenido {
    padding: 30px;
    text-align: center;
}
</style>

</head>

<body>

<div class="navbar">
    <a href="mostrar('ventas')">Ventas</a>
    <a href="mostrar('reportes')">Reportes</a>
    <a href="mostrar('noticias')">Noticias</a>
    <a href="/dashboard?usuario={{usuario}}&empresa={{empresa}}"Volver</a>
</div>

<div class="contenido" id="contenido">
    <h1>Bienvenido a Mercadeo</h1>
</div>

<script>
function mostrar(opcion){
    let contenido = document.getElementById("contenido");

    if(opcion === "ventas"){
        contenido.innerHTML = "<h2>Ventas</h2>;
    }

    if(opcion === "reportes"){
        contenido.innerHTML = "<h2>Reportes</h2>;
    }

    if(opcion === "noticias"){
        contenido.innerHTML = "<h2>Noticias</h2>;
    }
}
</script>

</body>
</html>
"""

talento_humano_html = """
<!DOCTYPE html>
<html>
<head>
<title>Talento Humano</title>

<style>
body {
    margin: 0;
    font-family: Arial;
}

.navbar {
    background-color: #34495e;
    padding: 10px;
    display: flex;
    gap: 20px;
}

.navbar a {
    color: white;
    text-decoration: none;
    cursor: pointer;
}

.navbar a:hover {
    text-decoration: underline;
}

.contenido {
    padding: 30px;
    text-align: center;
}
</style>

</head>

<body>

<div class="navbar">
    <a href="mostrar('Personal')">Personal</a>
    <a href="mostrar('Noticias')">Noticias</a>
    <a href="/dashboard?usuario={{usuario}}&empresa={{empresa}}"Volver</a>
</div>

<div class="contenido" id="contenido">
    <h1>Talento Humano</h1>
</div>

<script>
function mostrar(opcion){
    let contenido = document.getElementById("contenido");

    if(opcion === "Personal"){
        contenido.innerHTML = "<h2>Personal</h2>;
    }

    if(opcion === "Noticias"){
        contenido.innerHTML = "<h2>Noticias</h2>;
    }
</script>

</body>
</html>
"""

gerencia_html = """
<!DOCTYPE html>
<html>
<head>
<title>Gerencia</title>

<style>
body {
    margin: 0;
    font-family: Arial;
}

.navbar {
    background-color: #34495e;
    padding: 10px;
    display: flex;
    gap: 20px;
}

.navbar a {
    color: white;
    text-decoration: none;
    cursor: pointer;
}

.navbar a:hover {
    text-decoration: underline;
}

.contenido {
    padding: 30px;
    text-align: center;
}
</style>

</head>

<body>

<div class="navbar">
    <a href="mostrar('Indicadores')">Indicadores</a>
    <a href="mostrar('Noticias')">Noticias</a>
    <a href="/dashboard?usuario={{usuario}}&empresa={{empresa}}"Volver</a>
</div>

<div class="contenido" id="contenido">
    <h1>Gestion</h1>
</div>

<script>
function mostrar(opcion){
    let contenido = document.getElementById("contenido");

    if(opcion === "Indicadores"){
        contenido.innerHTML = "<h2>Indicadores</h2>;
    }

    if(opcion === "Noticias"){
        contenido.innerHTML = "<h2>Noticias</h2>;
    }
</script>

</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def login():
    error = ""
    empresas = ["Organizacion GyJ", "GyJ Ferreterias", "Almasa", "Colmena"]

    if request.method == "POST":
        username = request.form["usuario"]
        password = request.form["password"]
        empresa = request.form["empresa"]

        if username in usuarios:
            user_data = usuarios[username]

            if password == user_data["password"]:

                if user_data["empresa"] == "todas":
                    return redirect(url_for("dashboard", empresa=empresa, usuario=username))
                elif user_data["empresa"] == empresa:
                    return redirect(url_for("dashboard", empresa=empresa, usuario=username))
                else:
                    error = "No tienes acceso a esta empresa"
            else:
                error = "Contraseña incorrecta"
        else:
            error = "Usuario no existe"

        if username in usuarios:
            if usuarios[username]["empresa"] == "todas":
                empresas = ["Organizacion GyJ", "GyJ Ferreterias", "Almasa", "Colmena"]
            else:
                empresas = [usuarios[username]["empresa"]]

    return render_template_string(login_html, error=error, empresas=empresas)

@app.route("/dashboard")
def dashboard():
    usuario = request.args.get("usuario")
    empresa = request.args.get("empresa")
    return render_template_string(dashboard_html, usuario=usuario, empresa=empresa)

@app.route("/mercadeo")
def mercadeo():
    usuario = request.args.get("usuario")
    empresa = request.args.get("empresa")
    return render_template_string(mercadeo_html, usuario=usuario, empresa=empresa)

@app.route("/talento humano")
def talento_humano():
    usuario = request.args.get("usuario")
    empresa = request.args.get("empresa")
    return render_template_string(talento_humano_html, usuario=usuario, empresa=empresa)

@app.route("/gerencia")
def gerencia():
    usuario = request.args.get("usuario")
    empresa = request.args.get("empresa")
    return render_template_string(gerencia_html, usuario=usuario, empresa=empresa)
```
